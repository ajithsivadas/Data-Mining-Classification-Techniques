```python


import pandas as pd
import numpy as np

```

```python
#1) Loading Iris Data Set
#citation: https://pythonhealthcare.org/2018/04/14/61-machine-learning-the-iris-data-set/
from sklearn.datasets import *
iris = load_iris()     #loading the file
    
ir = pd.DataFrame(iris.data)#converting into dataframe

ir.columns = iris.feature_names #naming columns of the dataframe
print(ir.columns)

ir['CLASS'] = iris.target
print(ir)# ir is the dataframe
```

```python
#2) Split your dataset 70% for training, and 30% for testing the classifier. [2 points]
#citation: https://www.bitdegree.org/learn/train-test-split
from sklearn.model_selection import train_test_split
x = np.array(ir.iloc[:,[0,1,2,3]])
y = np.array(ir['CLASS'])
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 42)
print("Training data")
print(x_train)
print(y_train)
print("Testing data")
print(x_test)
print(y_test)
```

```python
#3) Select only 2 attributes for training and testing your model. [2 points]
#4) Use Euclidean distance. [3 points]
#5) Test the classifier with three different numbers for neighbors and record the results. [3 points]
#citation: https://www.kaggle.com/gilsousa/prediction-iris-dataset
from sklearn import neighbors
from sklearn.metrics import accuracy_score
#testing the classifier with neighbors = 3
np.random.seed(0)
knn = neighbors.KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(x_train[:,[1,2]],y_train)# selecting two attributes sepal width, petal length for training.
pred = knn.predict(x_test[:,[1,2]])# selecting two attributes sepal width, petal length for training
print("Predicted value for neighbors = 3 is",pred)
print("The accuracy for 3 neigbors is {} ".format(accuracy_score(y_test,pred)))
#testing the classifier with neighbor = 11
knn2 = neighbors.KNeighborsClassifier(n_neighbors=11, metric='euclidean')
knn2.fit(x_train[:,[1,2]],y_train)# selecting two attributes sepal width, petal length for training.
pred2 = knn2.predict(x_test[:,[1,2]])# selecting two attributes sepal width, petal length for training
print("Predicted value for neighbors = 11 is",pred2)
print("The accuracy for 11 neigbors is {} ".format(accuracy_score(y_test,pred2)))
#testing the classifier with neighbor = 18
knn1 = neighbors.KNeighborsClassifier(n_neighbors=18, metric='euclidean')
knn1.fit(x_train[:,[1,2]],y_train)# selecting two attributes sepal width, petal length for training.
pred1 = knn1.predict(x_test[:,[1,2]])# selecting two attributes sepal width, petal length for training
print("Predicted value for neighbors = 18 is",pred1)
print("The accuracy for 18 neigbors is {} ".format(accuracy_score(y_test,pred1)))
#testing the classifier with neighbor = 30
knn3 = neighbors.KNeighborsClassifier(n_neighbors=30, metric='euclidean')
knn3.fit(x_train[:,[1,2]],y_train)# selecting two attributes sepal width, petal length for training.
pred3 = knn3.predict(x_test[:,[1,2]])# selecting two attributes sepal width, petal length for training
print("Predicted value for neighbors = 30 is",pred3)
print("The accuracy for 30 neigbors is {} ".format(accuracy_score(y_test,pred3)))
```
